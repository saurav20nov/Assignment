Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of
Matplotlib.

Matplotlib is a popular data visualization library for Python that allows users to create a wide range of charts, graphs, and plots. It provides a set of functions for creating different types of plots, customizable styles, and tools for interacting with and modifying plots.

Matplotlib is used for data visualization in various fields like data science, engineering, finance, and more. It is a widely used library for creating visualizations because of its flexibility, ease of use, and extensive documentation.

The Pyplot module of Matplotlib provides a simple interface for creating different types of plots. Here are five plots that can be plotted using the Pyplot module of Matplotlib:

1. Line plot: A line plot is a simple plot that displays data points connected by lines. It is used to visualize trends or changes in data over time.

2. Scatter plot: A scatter plot is used to display the relationship between two variables. Each point on the plot represents a pair of values for the two variables.

3. Bar plot: A bar plot is used to compare the values of different categories. It displays bars of different heights, with each bar representing a category and the height representing the value.

4. Histogram: A histogram is used to visualize the distribution of a numerical variable. It displays the frequency of values in different intervals or bins.

5. Pie chart: A pie chart is used to display the relative sizes of different categories as slices of a pie. It is commonly used to show the composition of a dataset.







Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.

A scatter plot is a type of graph used to display the relationship between two variables. It displays data as a collection of points, with one variable on the x-axis and the other on the y-axis. Scatter plots are useful for identifying patterns and trends in data.

```python
import matplotlib.pyplot as plt
import numpy as np


np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))


plt.scatter(x, y)


plt.title('Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')


plt.show()
```

Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:
import numpy as np
For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])

The subplot() function in Matplotlib is used to create multiple plots within the same figure. This function allows users to divide the figure into multiple subplots or axes, which can then be used to display different data or different aspects of the same data.

```python
import matplotlib.pyplot as plt
import numpy as np


x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])


fig, axs = plt.subplots(2, 2)


axs[0, 0].plot(x1, y1)
axs[0, 0].set_title('Line 1')

axs[0, 1].plot(x2, y2)
axs[0, 1].set_title('Line 2')

axs[1, 0].plot(x3, y3)
axs[1, 0].set_title('Line 3')

axs[1, 1].plot(x4, y4)
axs[1, 1].set_title('Line 4')


fig.suptitle('Four Line Plots')


plt.show()
```

Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

A bar plot is a graphical representation of categorical data with rectangular bars of lengths proportional to the values they represent.

It is used to display and compare the values of different categories or groups. 

The bars can be vertical or horizontal, depending on the orientation of the plot.

```python
import numpy as np
import matplotlib.pyplot as plt


company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

plt.bar(company, profit)
plt.title('Profit by Company')
plt.xlabel('Company')
plt.ylabel('Profit (in millions)')
plt.show()


plt.barh(company, profit)
plt.title('Profit by Company')
plt.xlabel('Profit (in millions)')
plt.ylabel('Company')
plt.show()
```

Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

A box plot is a graphical representation of a dataset that provides a summary of the data's distribution.

It is also known as a box-and-whisker plot. The plot displays a box with lines extending from the top and bottom of the box, called whiskers.

The box represents the middle 50% of the data, with a line in the middle indicating the median value.

The whiskers represent the spread of the data beyond the box, typically extending to 1.5 times the interquartile range (IQR) from the top and bottom of the box. 

Any points outside the whiskers are considered outliers.





Box plots are used to quickly and easily visualize the distribution of a dataset and to identify outliers. They are particularly useful when comparing multiple datasets as they allow for easy comparison of the medians, ranges, and the spread of each dataset.


```python
import matplotlib.pyplot as plt
import numpy as np

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

data = [box1, box2]

fig, ax = plt.subplots()

ax.boxplot(data)

plt.show()
```
